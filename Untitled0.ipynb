```python
# passo 1 = Importar a base de dados
tabela = pd.read_excel('gdrive/My Drive/analise_dados/Vendas.xlsx')

```

```python
# Passo 2: Visualizar a base de dados
# Entender quais informações disponíveis 
# Descobrir as cagadas na base de dados
tabela = tabela.drop('Data', axis=1) # excluir linha 0 linha 1 coluna
display(tabela)
```

```python
# Passo 3 - tratamento na base de dados
# - valores que estão vazios
# - valores que são numeros mas o python acha que e texto
# - Informações inuteis
tabela['Valor Final'] = pd.to_numeric(tabela['Valor Final'], errors="coerce") # Trasnformar string em número 
# colunas vazias
tabela = tabela.dropna(how="all", axis=1) # all quando você quer excluir colunas tolmente vazias
# any é quando você quer excluir a coluna quando 1 dados esta vazio 

# Linhas vazias
tabela = tabela.dropna(how="any", axis=0) 
print(tabela.info()) # Informações sobre a tabela

```

```python
# Passo 4: Análise exploratoria -> Análise geral 

display(tabela['ID Loja'].value_counts())
display(tabela['ID Loja'].value_counts(normalize=True).map('{:.1%}'.format))
```

```python
# Passo 5: Olhando as colunas na base de dados - identificar o motivo dos clientes cancelaram 

for coluna in tabela.columns:
    print(coluna)
    grafico = px.histogram(tabela, x=coluna, color='Quantidade')
    grafico.show()
```
