```python
# passo 1 = Importar a base de dados
tabela = pd.read_excel('gdrive/My Drive/analise_dados/Vendas.xlsx')

```

```python
# Passo 2: Visualizar a base de dados
# Entender quais informa√ß√µes dispon√≠veis 
# Descobrir as cagadas na base de dados
tabela = tabela.drop('Data', axis=1) # excluir linha 0 linha 1 coluna
display(tabela)
```

```python
# Passo 3 - tratamento na base de dados
# - valores que est√£o vazios
# - valores que s√£o numeros mas o python acha que e texto
# - Informa√ß√µes inuteis
tabela['Valor Final'] = pd.to_numeric(tabela['Valor Final'], errors="coerce") # Trasnformar string em n√∫mero 
# colunas vazias
tabela = tabela.dropna(how="all", axis=1) # all quando voc√™ quer excluir colunas tolmente vazias
# any √© quando voc√™ quer excluir a coluna quando 1 dados esta vazio 

# Linhas vazias
tabela = tabela.dropna(how="any", axis=0) 
print(tabela.info()) # Informa√ß√µes sobre a tabela

```

```python
# Passo 4: An√°lise exploratoria -> An√°lise geral 

display(tabela['ID Loja'].value_counts())
display(tabela['ID Loja'].value_counts(normalize=True).map('{:.1%}'.format))
```

```python
# Passo 5: Olhando as colunas na base de dados - identificar o motivo dos clientes cancelaram 

for coluna in tabela.columns:
    print(coluna)
    grafico = px.histogram(tabela, x=coluna, color='Quantidade')
    grafico.show()
```
